In this project I have used support vector regressor to train and test boston dataset by using linear and rbf kernel and played with different values of c and gamma 

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
boston = datasets.load_boston()
```

```python
x = boston.data
y = boston.target
```

```python
x_train,x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
clf = svm.SVR(kernel="rbf")
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

we can clearly see that svr with rbf kernel is performing very bad. Let's play with the parameters like c(which isinversly proportional to strength of regularisation) and gamma (which inversly proportional to square of sigma) with the help of GridSearchCV

```python
clf1 = svm.SVR()
grid = {'C': [1e2,1e3, 5e3, 1e4, 5e4, 1e5],
       'gamma': [1e-3, 5e-4, 1e-4, 5e-3]}
res1 = GridSearchCV(clf, grid)
res1.fit(x_train, y_train)
res1.best_estimator_
```

```python
res1.score(x_test, y_test)
```

Let's try some more values of c and gamma 

```python
clf2 = svm.SVR()
grid = {'C': [1,1e1,1e2,1e3, 5e3, 1e4, 5e4, 1e5],
       'gamma': [1e-2,5e-3,5e-3,1e-3, 5e-4, 1e-4, 5e-3]}
res2 = GridSearchCV(clf2, grid)
res2.fit(x_train, y_train)
res2.best_estimator_
```

Still the same value of c and gamma are obtained  and so that'swhy no change is score also

```python
res2.score(x_test, y_test)
```

Let's try by taking the linear kernel 

```python
clf = svm.SVR(kernel="linear")
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

We can see that linear kernel is also performing well for the boston dataset
